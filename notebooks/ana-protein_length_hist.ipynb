# What is the distribution of protein length in the dataset?

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_dir = '../data/taxa/proteins/'
length_list = []
for i, f in enumerate(os.listdir(data_dir)):
    lengths = pd.read_csv(data_dir+f, usecols=[3], sep=';').values
    length_list.append(lengths)
```

```python
sns.histplot(np.vstack(length_list))
```

```python
data = np.vstack(length_list)
```
